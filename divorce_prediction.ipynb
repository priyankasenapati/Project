```python
import pandas as pd
from sklearn import preprocessing
from scipy.io import arff

# Read the divorce data
divorce_df = pd.read_csv("divorce_data/divorce.csv")
divorce_df
```

```python
#Check for NULLs in the value
nullcount = divorce_df.isnull().sum().sort_values(ascending=False)
print(nullcount)
```

```python
#Normalize the data
values = divorce_df.values #returns a numpy array
print(values)
min_max_scaler = preprocessing.MinMaxScaler()
values_scaled = min_max_scaler.fit_transform(values)
divorce_scaled_df = pd.DataFrame(values_scaled)
divorce_scaled_df
```
